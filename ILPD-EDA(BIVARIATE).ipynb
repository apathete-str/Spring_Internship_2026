```python
%mamba install pandas
import pandas as pd

import matplotlib.pyplot as plt
df = pd.read_csv("Indian Liver Patient Dataset (ILPD).csv", header=None)

df.columns = [
    'Age', 'Gender', 'Total Bilirubin', 'Direct Bilirubin', 
    'Alkaline Phosphotase', 'Alamine Aminotransferase', 
    'Aspartate Aminotransferase', 'Total Proteins', 'Albumin', 
    'Albumin and Globulin Ratio', 'Dataset']
```

Now , making

# THE PLOT OF THE CORELATION MATRICES : 
<hr>

# PEARSON CORELATION COEFFICIENT MATRIX :

```python
df['Gender-Number'] = df['Gender'].map({'Female': 0, 'Male': 1})
numeric_df = df.select_dtypes(include=['number'])

pearson_matrix = numeric_df.corr(method='pearson')
pearson_matrix
```

<hr>

# SPEARMAN CORRELATION COEFFICIENT MATRIX :

```python
spearman_matrix = numeric_df.corr(method='spearman')
spearman_matrix
```

<b><h3>1. Ignore Direct Bilirubin.
Keep Total Bilirubin.</h3></b>

Total Bilirubin & Direct Bilirubin (Pearson: 0.87 | Spearman: 0.96)

Reason: 
They are highly correlated.
Also, Direct Bilirubin is a fraction of Total Bilirubin,
<hr>

<b><h3>2. Ignore ALT. Keep AST Liver Enzyme.</h3><b>

AST & ALT Liver Enzymes (Pearson: 0.79 | Spearman: 0.77)

Reason: They have a strong positive linear relationship. When the liver is damaged, both of these leak into the blood together.
    <hr>

<h3><b> 3. Ignore Albumin . Keep Total Proteins. </b></h3>

Albumin & Total Proteins (Pearson: 0.78 | Spearman: 0.78)

Reason: They have a strong positive linear relationship. Albumin makes up the vast majority of Total Proteins, and is thus included in total number of proteins.

<hr>

Reason :
